```python
#Create training data file
create_images = False
data = []
pos_cnt = 1
neg_cnt = 1
all_data = pd.DataFrame()
folders = ["1", "2"]
for folder in folders:
    base_path = path.join(PATH, folder)
    print(base_path)
    ann = pd.read_csv(path.join(base_path, "via_export_csv.csv"))
    for file in ann["filename"].unique():
        print(file)
        sample_ann = ann[ann["filename"] == file]

        # Read all the annotations
        for ix, val in sample_ann.iterrows():
            coords = json.loads(val["region_shape_attributes"])
            if "x" not in coords.keys():
                continue
            if coords["x"] is None:
                continue
            center = [coords["y"] + (coords["height"] / 2), coords["x"] + (coords["width"] / 2)]
            sample_ann.loc[ix, "centery"] = center[0]
            sample_ann.loc[ix, "centerx"] = center[1]
        training_data = []

        # Split the image into sub images of size 40x40
        for _x in range(1, 99):
            for _y in range(1, 74):
#         for _x in range(1, 9):
#             for _y in range(1, 7):
                xstart = _x * 40
                xend = (_x + 1) * 40
                ystart = _y * 40
                yend = (_y + 1) * 40
                label = 0
                index = 0
                #print(val["centery"], val["centerx"], ystart, yend, xstart, xend)
                for ix, val in sample_ann.iterrows():
                    if "centery" not in val.keys():
                        continue
                    if (val["centery"] > ystart and 
                        val["centery"] < yend and
                        val["centerx"] > xstart and
                        val["centerx"] < xend):
                        label = 1
                        index = ix
                training_data.append([ystart, yend, xstart, xend, label, index, file, folder])
        df = pd.DataFrame(training_data, columns=["ys", "ye", "xs", "xe", "label", "index", "filename", "folder"])
        if create_images:
            df_true = df[df["label"] == 1]
            df_false = df[df["label"] == 0]
            print("creatin")
            while len(df_true) < len(df_false):
                df_true = df_true.append(df_true, ignore_index=True)
            # df_false = df[df["label"] == 0].sample(2)
            df_training = df_true.append(df_false)
            img = cv2.imread(path.join(base_path, file))
            for ix, val in df_true.iterrows():
                img_crop = img[val["ys"]: val["ye"], val["xs"]: val["xe"]]
                # print(img_crop.shape)
                cv2.imwrite(path.join(base_path, "positive", str(pos_cnt) + ".jpg"), img_crop)
                pos_cnt += 1
            for ix, val in df_false.iterrows():
                img_crop = img[val["ys"]: val["ye"], val["xs"]: val["xe"]]
                cv2.imwrite(path.join(base_path, "negative", str(neg_cnt) + ".jpg"), img_crop)
                neg_cnt += 1
        else:
            all_data = all_data.append(df)
all_data.to_csv(path.join(PATH,"all_cropped_40.csv"), index=False)
```

```python
#Create annotations by reshaping the data to 40x40 and randomly selecting the negative data
X = []
Y = []
for file in ann["filename"].unique():
    cnt1 = 0
    cnt0 = 0
    print(file)
    sample_ann = ann[ann["filename"] == file]
    img = cv2.imread(path.join(base_path, file))
    # Read all the annotations
    for ix, val in sample_ann.iterrows():
        coords = json.loads(val["region_shape_attributes"])
        x1 = coords["x"]
        x2 = x1 + coords["width"]
        y1 = coords["y"]
        y2 = y1 + coords["height"]
        for i in range(-3, 3):
            for j in range(-3, 3):             
                img_crop = img[y1+i:y2+i, x1+j:x2+j]
                X.append(cv2.resize(img_crop, (40, 40)))
                Y.append(1)
                cnt1 += 1
    for ix, val in sample_ann.iterrows():
        coords = json.loads(val["region_shape_attributes"])
        x1 = coords["x"]
        x2 = x1 + coords["width"]
        y1 = coords["y"]
        y2 = y1 + coords["height"]
        img_crop = img[y1:y2, x1:x2]
        img[y1:y2, x1:x2] = np.ones_like(img_crop)
    # img = cv2.imread(path.join(base_path, file))
    while not cnt0 == cnt1:
        y1 = random.randint(0, 2950)
        x1 = random.randint(0, 3950)
        y2 = y1 + 40
        x2 = x1 + 40
        img_crop = img[y1:y2, x1:x2]
        if len(np.where(img_crop==1)[0]) < 10:
            img_crop = img[y1:y2, x1:x2]
            X.append(img_crop.copy())
            Y.append(0)
            cnt0 += 1
            img[y1:y2, x1:x2] = np.ones_like(img_crop)
```

```python
folders = ["Set1"]
for folder in folders:
    base_path = path.join(PATH, folder)
    print(base_path)
    ann = pd.read_csv(path.join(base_path, "via_export_csv.csv"))  
    new_ann = []
    processed_files = []
    if not path.exists(path.join(base_path, "split")):
        os.makedirs(path.join(base_path, "split"))
    for ix in tqdm.tqdm_notebook(range(len(ann))):
        val = ann.iloc[ix]
        region_shape = json.loads(val["region_shape_attributes"])
        new_region_shape = deepcopy(region_shape)
        file = val["filename"]
        file_split = file.split(".")[0]
        if region_shape:
            if "x" in region_shape.keys():
                x = region_shape["x"]
                if x is not None:
                    y = region_shape["y"]
                    xbin = x // 400
                    ybin = y // 300
                    new_region_shape["x"] = x % 400
                    new_region_shape["y"] = y % 300
                    new_file = file_split + "_" + str(xbin) + "_" + str(ybin) + ".jpg"
                    new_ann.append([new_file, json.dumps(new_region_shape)])

        if file not in processed_files:
            if not file.startswith("2019_1206_122956_179"):
                continue
            img = cv2.imread(path.join(base_path, file))
            if img is None:
                continue
            for x in range(10):
                for y in range(10):
                    new_file = file_split + "_" + str(x) + "_" + str(y) + ".jpg"
                    new_image = img[y*300: (y+1)*300, x*400:(x+1)*400]
                    cv2.imwrite(path.join(base_path, "split", new_file), new_image)  
            processed_files.append(file)
    new_ann = pd.DataFrame(new_ann, columns=["filename", "region_shape_attributes"])
    new_ann.to_csv(path.join(base_path, "split", "via_export_csv.csv"), index=False)
```
